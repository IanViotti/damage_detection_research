```python
import os
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

sns.set_theme()
plt.rcParams["image.cmap"] = "Set2"
plt.rcParams['axes.prop_cycle'] = plt.cycler(color=plt.cm.Set2.colors)
```

```python
files = os.listdir()
list_files = []
for i in files:
    if 'xlsx' in i:
        file_1 = pd.read_excel(i, header = 2)
        list_files.append(file_1)
```

```python
data = np.reshape(list_files, (6,3))
```

```python
fig, (ax1, ax2, ax3, ax4, ax5) = plt.subplots(5,3, figsize = (15,20), dpi = 200)

# Dano 1 

ax1[0].plot(data[1][0]['X Value'], data[1][0]['Y Value(dBMag)'], label = '3x3 mm - Interface Damage')
ax1[1].plot(data[1][1]['X Value'], data[1][1]['Y Value(Imag)'], label = '3x3 mm - Interface Damage')
ax1[2].plot(data[1][2]['X Value'], data[1][2]['Y Value(Real)'], label = '3x3 mm - Interface Damage')

ax1[0].plot(data[0][0]['X Value'], data[0][0]['Y Value(dBMag)'], label = 'Undamaged')
ax1[1].plot(data[0][1]['X Value'], data[0][1]['Y Value(Imag)'], label = 'Undamaged')
ax1[2].plot(data[0][2]['X Value'], data[0][2]['Y Value(Real)'], label = 'Undamaged')



ax1[2].set_xlim(left = 0, right = 0.10)

ax1[0].set(xlabel = 'Frequency [Hz]', ylabel = 'Magnitude (real) [dB]', title = 'FRF (real)')
ax1[1].set(xlabel = 'Frequency [Hz]', ylabel = 'Magnitude (imaginary) [dB]', title = 'FRF (imaginary)')
ax1[2].set(xlabel = 'Time [s]', ylabel = 'Velocity [m/s]', title = 'Velocity Over Time')

ax1[0].legend()
ax1[1].legend()
ax1[2].legend()

# Dano 2 

ax2[0].plot(data[2][0]['X Value'], data[2][0]['Y Value(dBMag)'], label = '3x7 mm - 1st Ply Damage')
ax2[1].plot(data[2][1]['X Value'], data[2][1]['Y Value(Imag)'], label = '3x7 mm - 1st Ply Damage')
ax2[2].plot(data[2][2]['X Value'], data[2][2]['Y Value(Real)'], label = '3x7 mm - 1st Ply Damage')

ax2[0].plot(data[0][0]['X Value'], data[0][0]['Y Value(dBMag)'], label = 'Undamaged')
ax2[1].plot(data[0][1]['X Value'], data[0][1]['Y Value(Imag)'], label = 'Undamaged')
ax2[2].plot(data[0][2]['X Value'], data[0][2]['Y Value(Real)'], label = 'Undamaged')

ax2[2].set_xlim(left = 0, right = 0.10)

ax2[0].set(xlabel = 'Frequency [Hz]', ylabel = 'Magnitude (real) [dB]', title = 'FRF (real)')
ax2[1].set(xlabel = 'Frequency [Hz]', ylabel = 'Magnitude (imaginary) [dB]', title = 'FRF (imaginary)')
ax2[2].set(xlabel = 'Time [s]', ylabel = 'Velocity [m/s]', title = 'Velocity Over Time')

ax2[0].legend()
ax2[1].legend()
ax2[2].legend()

# Dano 3
ax3[0].plot(data[3][0]['X Value'], data[3][0]['Y Value(dBMag)'], label = '1.5x4.7 mm - Interface Damage')
ax3[1].plot(data[3][1]['X Value'], data[3][1]['Y Value(Imag)'], label = '1.5x4.7 mm - Interface Damage')
ax3[2].plot(data[3][2]['X Value'], data[3][2]['Y Value(Real)'], label = '1.5x4.7 mm - Interface Damage')

ax3[0].plot(data[0][0]['X Value'], data[0][0]['Y Value(dBMag)'], label = 'Undamaged')
ax3[1].plot(data[0][1]['X Value'], data[0][1]['Y Value(Imag)'], label = 'Undamaged')
ax3[2].plot(data[0][2]['X Value'], data[0][2]['Y Value(Real)'], label = 'Undamaged')

ax3[2].set_xlim(left = 0, right = 0.10)

ax3[0].set(xlabel = 'Frequency [Hz]', ylabel = 'Magnitude (real) [dB]', title = 'FRF (real)')
ax3[1].set(xlabel = 'Frequency [Hz]', ylabel = 'Magnitude (imaginary) [dB]', title = 'FRF (imaginary)')
ax3[2].set(xlabel = 'Time [s]', ylabel = 'Velocity [m/s]', title = 'Velocity Over Time')

ax3[0].legend()
ax3[1].legend()
ax3[2].legend()

# Dano 4
ax4[0].plot(data[4][0]['X Value'], data[4][0]['Y Value(dBMag)'], label = '1.5x5 mm - 3rd Ply Damage')
ax4[1].plot(data[4][1]['X Value'], data[4][1]['Y Value(Imag)'], label = '1.5x5 mm - 3rd Ply Damage')
ax4[2].plot(data[4][2]['X Value'], data[4][2]['Y Value(Real)'], label = '1.5x5 mm - 3rd Ply Damage')

ax4[0].plot(data[0][0]['X Value'], data[0][0]['Y Value(dBMag)'], label = 'Undamaged')
ax4[1].plot(data[0][1]['X Value'], data[0][1]['Y Value(Imag)'], label = 'Undamaged')
ax4[2].plot(data[0][2]['X Value'], data[0][2]['Y Value(Real)'], label = 'Undamaged')

ax4[2].set_xlim(left = 0, right = 0.10)

ax4[0].set(xlabel = 'Frequency [Hz]', ylabel = 'Magnitude (real) [dB]', title = 'FRF (real)')
ax4[1].set(xlabel = 'Frequency [Hz]', ylabel = 'Magnitude (imaginary) [dB]', title = 'FRF (imaginary)')
ax4[2].set(xlabel = 'Time [s]', ylabel = 'Velocity [m/s]', title = 'Velocity Over Time')

ax4[0].legend()
ax4[1].legend()
ax4[2].legend()

# Dano 5
ax5[0].plot(data[5][0]['X Value'], data[5][0]['Y Value(dBMag)'], label = '3x7 mm - Interface Damage')
ax5[1].plot(data[5][1]['X Value'], data[5][1]['Y Value(Imag)'], label = '3x7 mm - Interface Damage')
ax5[2].plot(data[5][2]['X Value'], data[5][2]['Y Value(Real)'], label = '3x7 mm - Interface Damage')

ax5[0].plot(data[0][0]['X Value'], data[0][0]['Y Value(dBMag)'], label = 'Undamaged')
ax5[1].plot(data[0][1]['X Value'], data[0][1]['Y Value(Imag)'], label = 'Undamaged')
ax5[2].plot(data[0][2]['X Value'], data[0][2]['Y Value(Real)'], label = 'Undamaged')

ax5[2].set_xlim(left = 0, right = 0.10)

ax5[0].set(xlabel = 'Frequency [Hz]', ylabel = 'Magnitude (real) [dB]', title = 'FRF (real)')
ax5[1].set(xlabel = 'Frequency [Hz]', ylabel = 'Magnitude (imaginary) [dB]', title = 'FRF (imaginary)')
ax5[2].set(xlabel = 'Time [s]', ylabel = 'Velocity [m/s]', title = 'Velocity Over Time')

ax5[0].legend()
ax5[1].legend()
ax5[2].legend()

fig.suptitle('Damaged and Undamaged Plates Modal Output Comparison \n', fontsize=20)

plt.tight_layout()

plt.savefig('modal_output_comp.pdf')
```
