## Reading FRF

The function `.read_ImFRF()` goes to exemperimental dataset and reads the imaginary part of the FRF from the plante called by input.

> The following plates are allowed: *placa sem dano, Placa dano 12 3x7, Placa dano 34 5x1,5, Placa dano inter 3x3, Placa dano inter 3x7, Placa dano inter 4,7x1,5*.


```python
plate = 'placa sem dano'

Im_FRF_plate = af.read_ImFRF(plate)
```

We have to define the mesuring points in the plate. 

For this case, we have 5 rows and 6 columns.

```python
rows = 5
columns = 6
n_points = rows * columns
```

## Setting the Im FRF amplitude for each node

Now we need to input the natural frequencies to obtain the ImFRF amplitude for each modal case.

> `.get_amplitude()` has *Im_FRF, w_exp and n_points, rows and mode* as input parameters and returns the formated amplitude for given mode on each row and collumn.

```python
#w_exp = (210.46, 360.80, 815.48, 1044.8) # resultados do ansys
w_exp = [256 , 488, 1180, 1488, 1568] # resultados do ensaio novo
```

```python
mode = 5 # Mode of vibration to extract
node_im = af.get_amplitude(Im_FRF_plate, w_exp, n_points, rows, mode)
```

# Plotting

Finally ploting plate vibration with surface plot.
